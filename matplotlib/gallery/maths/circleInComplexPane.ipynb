```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.ticker import FuncFormatter
```

```python
fig, ax = plt.subplots(figsize = (15, 15))

@FuncFormatter
def funcformatter(x, pos):
    if x == 0: 
        x = ''
    else:
        x = '${}i$'.format(round(x, 2))
    return x

# first circle
ax.set_aspect(1)
circle1 = mpl.patches.Circle((1, 0), radius = 1, color = 'red', fc = 'None')
ax.add_patch(circle1)

# second circle
circle2 = mpl.patches.Circle((0, 0), radius = 2, color = 'blue', fc = 'None')
ax.add_patch(circle2)

ax.spines.left.set_position(('data', 0))
ax.spines.bottom.set_position(('data', 0))
ax.spines[['top', 'right']].set_visible(False)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
ax.yaxis.set_major_formatter(funcformatter)
ax.tick_params(labelsize = 15)
ax.autoscale_view()

# legend (https://stackoverflow.com/questions/11423369/matplotlib-legend-circle-markers)
line1 = Line2D(range(1), range(1), color="white", marker='o', markerfacecolor="red", markersize = 20)
line2 = Line2D(range(1), range(1), color="white", marker='o',markerfacecolor="blue", markersize = 20)
ax.legend((line1, line2), ('$e ^ {2 \pi i \cdot t} + 1$', '$2 \cdot e ^ {\pi i \cdot t}$'), fontsize = 20, numpoints = 1, loc = 1)

plt.show()
```
